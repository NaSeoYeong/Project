```python
import pandas as pd

df_id = pd.read_csv('./data/hub_id.csv')

df_id['hub_device_id'] = df_id['hub_device_id'].fillna(-1)
df_id['hub_device_id'] = df_id['hub_device_id'].astype('int')
df_id['hub_device_id'] = df_id['hub_device_id'].apply(lambda x:None if x == -1 else x)
```

```python
df_hub = pd.read_csv('./data/hub_525065081.csv')
df_hub
```

```python
d1=[2179,1766,2220,1970,1702,2411,357,2552,2273,1416]

hub_list = df_id[df_id['sensor_device_id'].isin(d1)]
```

```python
hub_list
```

```python
import os


FILE_PATH = os.listdir('data/baby')

for file in FILE_PATH:
    
    sensor_id = int(file.split('_')[1].split('.')[0])
    hub_id = int(hub_list[hub_list['sensor_device_id'] == sensor_id]['hub_device_id'])
    
    sensor_df = pd.read_csv(f'./data/baby/{file}')
    sensor_df.rename(columns={'update':'updated'}, inplace=True)
    sensor_df['updated'] = pd.to_datetime(sensor_df['updated'])
    
    hub_df = df_hub[df_hub['device_id'] == hub_id]
    hub_df['updated'] = pd.to_datetime(hub_df['updated'])
    hub_df['updated'] = hub_df['updated'].dt.floor('S')
    
    result_df = pd.merge(sensor_df, hub_df, how='left', on='updated')
    
    result_df.to_csv(f'./data/senhub/hub_device_{sensor_id}.csv', index=False)
```
