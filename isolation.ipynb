### minmax-scale - IsolationForest 모델 구현

```python
import pandas as pd

df=pd.read_csv('./data/device_27.csv',index_col=['updated'])
df.index = pd.to_datetime(df.index)
df.reset_index(drop=False, inplace=True)
df
```

#### 1. min-max scaler

```python
from sklearn.preprocessing import MinMaxScaler

scaler_list  = []

device_list = df['device_id'].unique()

for id in  device_list:
    one_df = df[df['device_id']==id]
    result = MinMaxScaler().fit_transform(one_df[['temperature']])
    scaler_list.extend(result.reshape(-1))


df['temperature_diff'] = scaler_list
df
```

#### 2. data parshing

```python
import numpy as np

one_data = []
all_data = []

for id in df['device_id']:
    one_df = df[df['device_id'] == id]
    # all_result = np.array(one_df['temperature'])
    # all_data.append(all_result)

    for date in one_df['days'].unique():
        result_df = one_df[one_df['days'] == date]
        one_result = np.array(result_df['temperature'])
        one_data.append(one_result)
```

#### 3. IsolationForest

```python
from sklearn.ensemble import IsolationForest

model = IsolationForest()

for id in df['device_id']:
    one_df = df[df['device_id'] == id]

    for date in one_df['days'].unique():
        result_df = one_df[one_df['days'] == date]
        one_result = np.array(result_df['temperature'])
        one_data.append(one_result)
        
model.fit()
```
